```python
Q1 = md1['TotalVisits'].quantile(0.25)
Q2 = md1['TotalVisits'].quantile(0.5)
Q3 = md1['TotalVisits'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
UL = Q3 + 1.5 * IQR
LL = Q1 - 1.5 * IQR
```

```python
UL
```

```python
LL
```

```python
len(md1[(md1['TotalVisits']<LL) | (md1['TotalVisits']>UL)])
```

```python
md3 = md1[(md1['TotalVisits']>LL) & (md1['TotalVisits']<UL)]
len(md3)
```

```python
print("Skewness of Original No Total Visits",md1['TotalVisits'].skew())
print("Skewness of IQR Cleaned Total Visits",md3['TotalVisits'].skew())
```

```python
print("Kurtosis of of Original No Total Visits",md1['TotalVisits'].kurt())
print("Kurtosis of  IQR Cleaned Total Visits",md3['TotalVisits'].kurt())
```
